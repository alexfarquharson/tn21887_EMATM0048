```python
class Board():
    """The Board class
    """
    
    def __init__(self, dimension):
        self.dimension = dimension
        self.table = self.create_board(dimension)
        self.player_turn = "1"
        self.show()
        


    def create_board(self, dimension):
        assert type(dimension) == int, "Please input an integer for board size"
        assert dimension > 3, "Board size must be greater than 3"
        
        top_row = [" ","_"] * dimension
        top_row.append(" ")
        top_row.append("\n")
        top_row = np.array(top_row)
        
        base_row = np.array([["|","_","|","\n"]])

        border_pos = (dimension-1)*[2]
        bottom_pos = []
        for x in range(dimension-1):
            bottom_pos.append(x+3)

        base_row = np.insert(base_row,border_pos,"Â¦")
        base_row = np.insert(base_row,bottom_pos,"_")

        for x in range(dimension):
            if x ==0:
                table = np.append(top_row,base_row)
            else:
                table = np.append(table,base_row)
            x+=1

        table = table.reshape(dimension+1,2*(dimension+1))
        
        table = self.insert_players(table,dimension)

        return table

    @staticmethod
    def insert_players(table,dimension):
    # insert players on board (customise character symbols?) add on @ symbol?
        table[1,1] = "1"
        table[dimension,2*dimension-1] = "2"

        return table
    
    def show(self):
        """Function to print the board table"""
        
        for y in range(self.table.shape[0]):
            for x in self.table[y,:]:
                print(x, end="")
                
                

    def move(self):
        
        move_dictionary = {"w":self.up, "a":self.left, "s":self.down, "d":self.right}
        direction = input(f'Player {self.player_turn} move: ').lower()
#         assert direction in move_dictionary.keys, ""
        try:
            legal = list(map(lambda x: move_dictionary[x],direction))[0]()
            
        except KeyError:
            print('KEY ERROR')
            legal = False
       
        return legal

    
    

    def find_player(self):
        row, col = np.where(self.table == self.player_turn)
        return row[0], col[0]
    
    def up(self):
        
        row, col = self.find_player()
        if row == 1:
            return False
        else:
            if self.table[row-1, col] == "_":
                self.table[row, col] = "X"
                self.table[row-1, col] = self.player_turn
                return True     
            else:
                return False
            
    def left(self):
        
        row, col = self.find_player()
        if col == 1:
            return False
        else:
            if self.table[row, col-2] == "_":
                self.table[row, col] = "X"
                self.table[row, col-2] = self.player_turn
                return True     
            else:
                return False     
            
    def right(self):
        
        row, col = self.find_player()
        if col == 2*self.dimension-1:
            return False
        else:
            if self.table[row, col+2] == "_":
                self.table[row, col] = "X"
                self.table[row, col+2] = self.player_turn
                return True     
            else:
                return False     

    def down(self):
        
        row, col = self.find_player()
        if col == self.dimension+1:
            return False
        else:
            if self.table[row+1, col] == "_":
                self.table[row, col] = "X"
                self.table[row+1, col] = self.player_turn
                return True     
            else:
                return False     
            

    def __main__(self):
        
#         print(f"\n Player {str(self.player_turn)}'s turn: \n ")

        legal = True
        while legal == True:
            legal = self.move()
            if legal == True:
                self.show()
                self.player_turn = str((int(self.player_turn))%2 +1)
        
        print(self.player_turn, "lost")
```
